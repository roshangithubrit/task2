```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("/content/archive (1).zip")
df
```

```python

legit = df[df.Class == 0 ]
fraud = df[df.Class == 1]
```

```python

df.groupby('Class').mean()
```

```python
pd_sample = df.sample(n= 4920)
```

```python
new = pd.concat([pd_sample, fraud],axis =0)
new
```

```python
new['Class'].value_counts()
```

```python
x = new.drop(columns = 'Class', axis=1)
```

```python
y = new['Class']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
model = SVC(kernel='linear')
```

```python
model.fit(x_train, y_train)
```

```python
X_train_prediction = model.predict(x_test)
data_accquary = accuracy_score(X_train_prediction,y_test)
```

```python
print(" The data_accquary is ", data_accquary)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
linear_model = LogisticRegression()
```

```python
linear_model.fit(x_train,y_train)
```

```python
X_train_prediction1 = linear_model.predict(x_test)
data_accquary1 = accuracy_score(X_train_prediction1,y_test)
```

```python
print(" The data_accquary is ", data_accquary1)
```

```python

from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
rf_X_test_prediction  = rf.predict(x_test)
data_accquary91 = accuracy_score(rf_X_test_prediction,y_test)
```

```python
print(" The data_accquary is ", data_accquary91)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python

clf.fit(x_train,y_train)
```

```python
clf_X_test_prediction  = clf.predict(x_test)
data_accquary_rf  = accuracy_score(clf_X_test_prediction,y_test)
```

```python
print(" The data_accquary is ", data_accquary_rf)
```
